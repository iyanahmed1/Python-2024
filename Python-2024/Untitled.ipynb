```python
# creating a 3d contour plot using python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#create a mesh grid of x and y values
x = np.linspace(-5, 5,100)
y = np.linspace(-5, 5,100)
X, Y = np.meshgrid(x,y)
```

```python
#define a function to calculate the z values (height) based on x and y
def f(x,y):
    return np.sin(np.sqrt(x**2 + y**2))
#value of z
Z = f(X, Y)
```

```python
# create a three dimensional contour plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111,projection= '3d')
contour = ax.contour3D(X, Y, Z, 50, cmaps='viridis')

#add label to the chart
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

fig.colorbar(contour, ax=ax, label = 'Z values')

#show the plot
plt.show()
```
